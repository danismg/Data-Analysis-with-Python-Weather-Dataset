## Find out all the Null Values in the data

```python
data.isnull() # return by boolean
data.isnull().sum()
```

## Rename the column name 'Weather' if the dataframe to 'Weather Condition'

```python
data.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True)
# Note :
# must be 'inplace = true' for the dataframe to change.
data
```

## What the mean 'Visibility'?

```python
data.Visibility_km.mean()
```

## What is the Standar Deviation of 'Pressure' in this data?

```python
data.Press_kPa.std()
```

## What is the Variance of 'Relative Humidity' in this data?

```python
data['Rel Hum_%'].var()
```

## Find all instances when 'Snow' was recorded.

```python
# Only Snow in Weather Condition
data[data['Weather Condition']== 'Snow']
```

```python
# Contains Snow
data[data['Weather Condition'].str.contains('Snow')].head()
```

## Find all instances when 'Wind Speed is above 24' and 'Visibility is 25'

```python
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

## What is the Mean value of each column against each 'Weather Condition'?

```python
data.groupby('Weather Condition').mean()
```

## What is the Minimum & Maximum value of each column againts each 'Weather Condition'?

```python
data.groupby('Weather Condition').max()
```

```python
data.groupby('Weather Condition').min()
```

## Show all the records where Weather Condition is fog

```python
data[data['Weather Condition']=='Fog']
```

## Find all instances when 'Weather is Clear' or 'Visibility above 40'

```python
data[(data['Weather Condition'] == 'Clear') | (data['Visibility_km']>40)]
```

## Find all instances when : 
'Weather is Clear' AND 'Relative Humidity is greater than 50'  <br>
OR<br>
'Visibility is above 40'

```python
data[((data['Weather Condition'] == 'Clear') & (data['Rel Hum_%'] > 50)) | (data['Visibility_km'] > 40)]
```
